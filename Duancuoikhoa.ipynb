```python
class Manage_item:
    def __init__(self, item, quantity):
        self.item = item
        self.quantity = quantity
    def __str__(self):
        return f"Name: {self.item}, Quantity: {self.quantity}"   
    
from tkinter import *
from tkinter import messagebox
import time
import csv
import os

# tạo cửa sổ windows
window = Tk()
window.title("Storage Tool")
window.geometry("730x500")
lb = Listbox(window,width = 120, height = 25)
# tao 1 list trống để lát sau input dữ liệu vào list này.
Item_list = []

def input_item():
    name_value = name.get()
    quantity_value = quantity.get()
    named_tuple = time.localtime() # get struct_time
    time_string = time.strftime("%m/%d/%Y, %H:%M:%S", named_tuple)
    
    if name_value == "" or quantity_value == "":
        messagebox.showerror("Item","Please input item and quantity")
    
    else:
        manage_value = Manage_item(name_value, int(quantity_value))
        Item_list.append(manage_value)
        print([Item.item for Item in Item_list],[Item.quantity for Item in Item_list], "Input time:",time_string)
        with open("Result.csv","a", newline="") as f:
            write = csv.writer(f)
            write.writerow(Item_list)
        show_item()
        messagebox.showinfo("Item",f"Input {name_value} successfully")
        

def output_item():
    named_tuple = time.localtime() # get struct_time
    time_string = time.strftime("%m/%d/%Y, %H:%M:%S", named_tuple) 
    Item_out = name.get()
    quantity_out = quantity.get()
    check = False
    for i in range(len(Item_list)):
        if Item_list[i].item == Item_out:
            Item_remain = int(Item_list[i].quantity) - int(quantity_out)
            Item_list[i].quantity = Item_remain
            with open("Result.csv","a",newline = "") as f:
                write = csv.writer(f)
                write.writerow(Item_list)
            
            messagebox.showinfo("Output Item",f"Item: {Item_list[i].item} remain {Item_remain}")
            print([Item.item for Item in Item_list], [Item.quantity for Item in Item_list], "Output time: ", time_string)
            check = True
            break
    if check == False:
        messagebox.showerror("Cannot Output Item","Item not change")
            
def search_item():
    Search_name = search_name.get()
    check = False
    
    for i in range(len(Item_list)):
        if Item_list[i].item == Search_name:
            messagebox.showinfo("Item found",f"Item status: Item: {Item_list[i].item}, Quantity: {Item_list[i].quantity}")
            check = True
            break
    if check == False:
        messagebox.showerror("Item Not Found", "Cannot find the Item")
        
        
def update_item():
    update_Item = name.get()
    update_Quantity = quantity.get()
    
    named_tuple = time.localtime() # get struct_time
    time_string = time.strftime("%m/%d/%Y, %H:%M:%S", named_tuple) 
    check = False
    if update_Item == "" or update_Quantity == "":
        messagebox.showerror("Item","Please enter name and quantity item")
    
    else:
        UPDATE_ITEM = Manage_item(update_Item, int(update_Quantity))

        for i in range(len(Item_list)):
            if Item_list[i].item == update_Item:
                Item_list[i].quantity = update_Quantity
                with open("Result.csv","a",newline = "") as f:
                    write = csv.writer(f)
                    write.writerow(Item_list)
                
                print([Item.item for Item in Item_list],[i.quantity for i in Item_list],"Update time:",time_string)
                messagebox.showinfo("Update Item",f"Update {update_Quantity} of {update_Item} successfully !")
                check = True
                break
        if check == False:
            messagebox.showerror("Cannot find Item","Cannot find the Item in Storage")

def delete_item():
    delete_name = search_name.get()
    named_tuple = time.localtime() # get struct_time
    time_string = time.strftime("%m/%d/%Y, %H:%M:%S", named_tuple)
    check = False
    
    for i in range(len(Item_list)):
        if Item_list[i].item == delete_name:
            del Item_list[i]
            with open("Result.csv","a",newline = "") as f:
                write = csv.writer(f)
                write.writerow(Item_list)
            print([Item.item for Item in Item_list],[Item.quantity for Item in Item_list], "Delete time:",time_string)
            messagebox.showinfo("Delete Item",f"Delete {delete_name} successfully !")
            check = True
            break
    if check == False:
        messagebox.showinfo("Delete Book","Cannot Find Item")

def show_item():
    storage = []
    lb.delete(0,END)
    
    with open("Result.csv","r") as f:
        reader = csv.reader(f)
        for i in reader:
            storage.append(i)
        for data in storage:
            lb.insert(END,data)
    return storage
    
    

# tạo các tên label hiển thị trên windows

lbl_name = Label(window, text="Name   :")
lbl_quantity = Label(window, text="Quantity    :")

# tạo các ô trống để điên thông tin (entry):

name = Entry(window)
quantity = Entry(window)
search_name = Entry(window)

# tạo các nút bấm

btn_input = Button(window, text ="Input ", command = input_item)
btn_output = Button(window, text = "Output", command = output_item)
btn_search = Button(window, text = "Search",command = search_item)
btn_delete = Button(window, text = "Delete",command = delete_item)
btn_update = Button(window ,text = "Update",command = update_item)
btn_show = Button(window,text = "Show ",command = show_item)

# gán các label, entry và button lên nền window

lbl_name.grid(row = 0, column = 0)
name.grid(row = 0, column = 1)
lbl_quantity.grid(row = 0, column = 2)
quantity.grid(row = 0 , column =3 )
btn_input.grid(row = 3 , column = 0)
btn_output.grid(row = 3 , column = 1)
btn_update.grid(row = 3, column = 2)
btn_delete.grid(row = 3 ,column = 3)
btn_search.grid(row = 4, column = 0)
search_name.grid(row = 4 , column =1)
btn_show.grid(row = 4 , column = 2)
lb.grid(row = 5 , columnspan = 4)
window.mainloop()
```
