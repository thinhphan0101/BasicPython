```python

class Book:
    def __init__(self, name, genre, price ):
        self.name = name
        self.genre = genre
        self.price = price
    def __str__(self):
        return f"Name: {self.name}, Genre: {self.genre}, Price: {self.price}"



from tkinter import *
from tkinter import messagebox


# tạo cửa số windows
window = Tk()
window.title("Book")

book_list = []

def add_book():
    name_value = name.get()
    genre_value = genre.get()
    price_value = price.get()
    
    if name_value == "" or genre_value == "" or price_value == "":
        messagebox.showerror("Book","Please input name genre price")
    else:
        new_book = Book(name_value, genre_value, int(price_value))
        book_list.append(new_book)
        
        print([book.name for book in book_list],[book.genre for book in book_list],[book.price for book in book_list])
        messagebox.showinfo("Book",f"Add {name_value} successfully!")
           
def delete_book():
    delete_name = search_name.get()
    check = False    
    for i in range(len(book_list)):
        if book_list[i].name == delete_name:
            del book_list[i]            
            print([book.name for book in book_list],[book.genre for book in book_list],[book.price for book in book_list])
            messagebox.showinfo("Delete Book", f"Delete {delete_name} successfully!")
            check = True
            break
        
        if check == False:
            messagebox.showinfo("Delete Book","Cannot Find Book")
        
def search_book():
    search_data = search_name.get()
    check = False
    for i in range(len(book_list)):
        if book_list[i].name == search_data:
            messagebox.showinfo("Search Book", book_list[i])
            check = True
            break      
        if check == False:
            messagebox.showinfo("Search Book","Cannot Find Book")
            
def update_book():
    update_Book = name.get()
    update_price = price.get() 
    check = False
    if update_Book == '' or update_price == '':
        messagebox.showerror('Updated Price Error', 'Please enter name and updated price')
    else:
        for i in range (len(book_list)):
            if book_list[i].name == update_Book:
                book_list[i].price = update_price
                print([book.name for book in book_list],[book.genre for book in book_list],[book.price for book in book_list])
                messagebox.showinfo('Price Updated', f'The price of the book {update_Book} has been updated to {update_price}$')
                check = True
                break
        if check == False:
                messagebox.showerror('Cannot Find Book', 'Cannot find the book in Database')



            

# tạo các tên hiển thị trên nền cửa số windows
lbl_name = Label(window, text="Name:")
lbl_genre = Label(window, text="Genre:")
lbl_price = Label(window, text ="Price: ")


# tạo các ô trống để điền thông tin
name = Entry(window)
genre = Entry(window)
price = Entry(window)
search_name = Entry(window)


# tạo các nút bấm
btn_add = Button(window, text = "Add", command=add_book)
btn_delete = Button(window, text="Delete",command = delete_book)
btn_search = Button(window, text="Search",command = search_book)
btn_Update = Button(window, text = "Update", command = update_book)


# gán các label, entry và button lên nền window
lbl_name.grid(row = 0, column = 0)
name.grid(row=0, column=1)
lbl_genre.grid(row = 1, column = 0)
genre.grid(row = 1 , column = 1)
lbl_price.grid(row = 2 , column = 0)
price.grid(row = 2, column = 1)
btn_add.grid(row = 3, column= 0 )
btn_delete.grid(row =3 , column =1)
btn_Update.grid(row = 3 , column = 2)
search_name.grid(row = 4, column = 0)
btn_search.grid(row = 4 , column = 1)


window.mainloop()
```
